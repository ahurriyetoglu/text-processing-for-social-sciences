## 1.3 Social Sciences and Topic Modeling <a class="anchor" id="one-three"></a>
 
**Political science:** Topic modeling can be used to **analyze political speeches, news articles, or other texts related to politics in order to identify the key themes or topics that are being discussed.** This can help researchers to understand how political discourse is changing over time, or to identify the key issues that are driving political debates.

[Greene, D., & Cross, J. P.](https://www.jstor.org/stable/26563293): This paper explored different debates in the European Parliment using Dynamic Topic Modeling.

**Sociology:** Topic modeling can be used to analyze **social media data or other online forums in order to identify the key topics or themes that people are discussing.** This can help researchers to understand how people are talking about different social issues, or to identify emerging trends or patterns in social discourse.


[Nelson](https://journals.sagepub.com/doi/pdf/10.1177/0049124117729703): The paper used topic modeling as part of a three-step computational grounded theory method to analyze qualitative text data and construct abstract conclusions about social reality, as demonstrated in an example analyzing the political logics underlying women's movement fields in New York City and Chicago from 1865 to 1975.


**Psychology:** Topic modeling can be used to analyze **patient data or other types of text data in order to identify the key themes or topics** that are associated with different psychological conditions or experiences. This can help researchers to better understand how different psychological conditions are related, or to identify new patterns or trends in psychological research.

[Franz](https://nocklab.fas.harvard.edu/files/nocklab/files/franz.pdf): The study used topic modeling to identify 15 latent topics related to **SITBs (Self-Injurious Thoughts and Behaviors)** and found gender and age differences in their discussion patterns in online forums.

**Economics:** Topic modeling can be used to analyze economic data, such as news articles or financial reports, in order to identify the key economic trends or topics that are being discussed. This can help researchers to better understand how economic policies and events are affecting different sectors of the economy, or to identify emerging trends or patterns in economic discourse.

[Priola et.al.](https://www.aimspress.com/article/id/62cfeb90ba35de7bde6cf733): This paper used topic modeling (Latent Dirichlet Allocation) to uncover latent topics in central banks' speeches and releases and then built a topic-specific sentiment index based on dictionary techniques to summarize the results in a topic-weighted Central Bank Sentiment Index (CBSIw) for several central banks.

Then came other topic models: 

**Dynamic topic modeling** is a variation of LDA that allows topics to change over time, so that the topics in a document or corpus can be analyzed as they evolve over time. 

*For example: Analyze the change in topics over time in a corpus of political addresses of presidents.*
 
[Exploring the UN General Debates with Dynamic Topic Models, Luke Lefebure, Towards Data Science](https://towardsdatascience.com/exploring-the-un-general-debates-with-dynamic-topic-models-72dc0e307696)
[Github Repo](https://github.com/llefebure/un-general-debates)

DTM is an extension of LDA that models the evolution of topics over time. It assumes that the topics change gradually over time, and that the content of a document is a mixture of topics that are active during the time period in which the document was written. DTM allows you to estimate **the topic-word and topic-document distributions** for each time period separately, as well as the **transition probabilities** between the topics from one time period to the next. This allows you to analyze how the topics change over time and which topics become more or less important in different time periods. 

Going back at the example, one could use DTM to model the change in topics over the years of different presidents. By estimating the topic distributions for each year, one could analyze how the topics discussed in political addresses changed over time and whether there were any significant shifts in the topics discussed during different presidencies.

## 2.1 Latent Dirichlet Allocation (LDA) <a class="anchor" id="two-one"></a>

[A Beginner's Guide to LDA, Kulshrestha](https://towardsdatascience.com/latent-dirichlet-allocation-lda-9d1cd064ffa2)

**Latent Dirichlet Allocation (LDA) is a generative probabilistic model, which means that it generates data according to a probability distribution.** In the case of LDA, the data it generates are documents, and the probability distribution it uses is a mixture of topics.

In LDA, each document is represented as a mixture of topics, where a topic is a probability distribution over words. 

For example, a topic might be about sports, and the probability distribution might assign high probabilities to words like "football," "basketball," and "tennis." 

Another topic might be about politics, and the probability distribution might assign high probabilities to words like "president," "government," and "elections."

The goal of LDA is to learn the topics and their associated probability distributions from a corpus of documents. It does this by assuming that each document is generated in the following way:

<span style = "color:green" > **Each document can be described by the probabilistic distribution of topics.** </span>

<span style = "color:green" > **Each topic can be described by the probabilistic distribution of words.**  </span>


The choice of the number of topics is a hyperparameter of the model, and it **needs to be set before training the model.** One way to determine the optimal number of topics is to use a metric like **coherence**. Topic model doesn't tell you how many topics there might be in your corpus (except GSDMM) instead you have to tell the topic model.


**One limitation of LDA is that it assumes that each word in a document is generated independently of the other words.** This assumption may not hold true in practice, as words in a document are often related to each other. To address this limitation, extensions to LDA have been proposed, such as Hierarchical Dirichlet Process (HDP) LDA and Dynamic Topic Models (DTM), which model the dependencies between words and topics.

In summary, LDA is a statistical model used for topic modeling that represents each document as a mixture of topics, where each topic is a probability distribution over words.

## 2.2 Gibbs Sampling for Dirichlet Multinomial Mixture (GSDMM) <a class="anchor" id="two-TWO"></a>

GSDMM stands for "Gibbs Sampling for Dirichlet Multinomial Mixture." It is a probabilistic topic modeling algorithm that is used to discover the latent themes or topics in a corpus of text documents.

**The algorithm assumes that each word in a document is generated by a single topic. This is a simplifying assumption that allows the algorithm to be computationally efficient and scalable.**

 A high value of alpha leads to more **concentrated distributions**, meaning that documents are likely to have fewer topics, while a low value of alpha leads to more diffuse distributions, meaning that documents are likely to have more topics.

 A high value of beta (eta) leads to more **concentrated distributions**, meaning that topics are likely to have fewer words, while a low value of beta leads to more diffuse distributions, meaning that topics are likely to have more words.

**The GSDMM algorithm has several advantages over other topic modeling algorithms.** One advantage is that it is computationally efficient and scalable, making it suitable for large-scale text analysis. 

**Another advantage is that it does not require the number of topics to be specified in advance**, as is the case with other topic modeling algorithms such as Latent Dirichlet Allocation (LDA). Instead, the GSDMM algorithm learns the number of topics from the data.

In LDA, documents are considered to be a mixture of topics. Usually, short text only deals with one topic so this assumption is not reliable anymore; Shorter text means fewer data to rely on during the LDA inference steps.
Well, the Gibbs Sampling Dirichlet Mixture Model (GSDMM) is an “extended” LDA algorithm, that makes the initial assumption: 1 topic is 1 document.

The words within a document are generated using the same unique topic, and not from a mixture of topics as it was in the original LDA.


GSDMM is a good choice for short text topic modeling. ([Reference](https://www.kaggle.com/code/ptfrwrd/topic-modeling-guide-gsdm-lda-lsi))

GSDMM (Gibbs Sampling Dirichlet Mixture Model) is primarily a **clustering algorithm**, not a topic modeling algorithm. The algorithm uses a probabilistic approach to assign data points to clusters based on their probability of belonging to each cluster, using Gibbs sampling to update the assignment probabilities.

While GSDMM is not a topic modeling algorithm per se, it can be used for **clustering text data based on the topics discussed in the text**. In this context, GSDMM can be seen as a topic-based clustering algorithm, where the clusters represent the different topics discussed in the text. The algorithm has been shown to be effective in clustering short text data, such as tweets and forum posts, where traditional topic modeling algorithms like Latent Dirichlet Allocation (LDA) may not perform well due to the sparsity of the data.



### 2.2.1 How does GSDMM differ from LDA?
1. LDA is a **generative probabilistic model** that assumes each document is a mixture of multiple topics, while each topic is a distribution over words. GSDMM is a **non-parametric clustering algorithm** that assigns each document to a single topic, based on the observed frequency of words in the document.

2. **In LDA, the number of topics is typically set in advance, while GSDMM infers the number of topics from the data.**

3. **GSDMM is generally faster than LDA**, because it assigns each document to a single topic instead of estimating a distribution over multiple topics.

4. **Both LDA and GSDMM can produce accurate topic models, but the performance can depend on the specific data set and task at hand.** In general, LDA is more widely used and better suited for tasks that require a finer-grained analysis of topics, while GSDMM can be more useful for tasks that require quick and scalable topic modeling.

## 2.3 Non-negative matrix factorization (NMF) <a class="anchor" id="two-three"></a>

NMF has been widely used in machine learning and data analysis, particularly in the field of unsupervised learning. It can be used for a variety of tasks, including dimensionality reduction, clustering, and feature extraction. NMF is particularly useful in cases where the data is inherently non-negative, such as in image processing and text analysis.

Some pros and cons of non-negative matrix factorization (NMF) specifically in the context of short text topic modeling:

**Pros:**

**Interpretable topics:** NMF can produce basis vectors and coefficients that are easier to interpret in the context of short texts. This is because the basis vectors may represent specific features or patterns in the text, such as keywords or phrases, and the coefficients may indicate the importance of these features in each document.

**Robustness to noise:** NMF can be more robust to noise and outliers in short text data than other topic modeling techniques, such as Latent Dirichlet Allocation (LDA).

**Faster computation:** NMF can often be computed more quickly than other topic modeling techniques, making it more practical for processing large volumes of short text data.

**Cons:**

**Limited coverage of topics:** NMF may not be able to capture as many distinct topics in short text data as other techniques, such as LDA, which can result in less comprehensive coverage of the topics in the data.

**Requires pre-processing:** NMF may require pre-processing of the short text data, such as stemming or stopword removal, to improve the quality of the topics it produces.

## 2.4 Do we need preprocessing for topic modeling?  <a class="anchor" id="two-four"></a>


Topic models work by identifying and grouping words that ***co-occur*** into “topics.” 

According to David Blei (LDA), two key presumptions underlie Latent Dirichlet allocation (LDA) topic modeling: 

(1) There are a predetermined number of word use patterns, or clusters of words that **frequently** appear together in written works. These are topics. 

(2) The subjects are presented in various ways in each text in the corpus.


Topic models cannot **"read"** the statement since they lack **any genuine semantic understanding** of the terms. Topic models instead rely on **mathematics**. Statistics suggest that the tokens or words that frequently appear together are connected. That said, if unimportant but frequently used phrases are utilized, the model is vulnerable to "noise," or mistakenly recognizing patterns of cooccurrence.

Unlike some deep learning models which don't require preprocessing, to ensure that the topic model detects interesting or significant patterns rather than noise, we should perform some preprocessing techniques.

**Preprocessing Tips for Topic Modeling**

[Preparing your data for Topic Modeling](https://publish.illinois.edu/commonsknowledge/2017/11/16/preparing-your-data-for-topic-modeling/)

A rule of thumb might be to remove every "unimportant" information as well as information that could potentialy bias your topics. 

These might include:

1. **Removing punctuations:** Without doing this step, commas will start to appear in every one of your results. 

2. **Removing stopwords:** If one doesn't exclude them, they will show up in your topics. Although you should be cautious about what stop words you eliminate, this might have a tendency to bias one's results. Stop-words should never be removed without first verifying them, since doing so may result in the unintentional omission of crucial information.

3. **Eliminating overused phrases:** Eliminate phrases that were only found in one's data. For instance, the phrase "rt" appeared frequently in my tweets since they were retweets etc. 

4. **Removing @mentions and URLs:** These might not show up on your topics because they don't occur frequently, but if you're dealing with a data that has a lot of retweets, one might want to consider removing them or remove the retweets completely. 

5. **Removing hashtags:** If they show up on your most frequent words sample, one might want to remove them. If one scraped tweets through a hastag list, they will cloud your results. If these aren't the case and you have a variety of hastags, you might keep them by just removing the ***"#"***. Although some hashtags don't make sense like ***#ulkemdegocmenistemiyorum*** ; some might contribute to the topics like ***#yagmacilar***. I guess it depends on your work. 

**IMPORTANT**: If one is using a context specific data (f.e. corpus of tweets scraped using a set of keywords etc.), one might think of removing the keyword list as well as the most occuring words. This can be a quantitative (print out the most frequent 30 words in the corpus ) as well as a qualitative process (deciding which most frequent words would create noise to see the diversity of real topics). If you left the most frequent words in the context specific corpus, your topics will look ***very*** similar thus missing one's mark. 

6. There is evidence that lemmatization appears to better the performance of topic model algorithms in English. ([Reference](https://cdn2.hubspot.net/hubfs/2396105/Topic%20modeling.pdf?t=1497538739976))

Making N-Grams

Bigrams are created for topic models to capture the context and co-occurrence of words in a text corpus. In a corpus, certain words tend to appear together more often than others. These word pairs, called bigrams, can be important for understanding the meaning of a text.

Topic models aim to identify latent topics in a corpus, which are groups of words that frequently co-occur in the same documents. Bigrams can help topic models by identifying meaningful combinations of words that should be treated as a single unit.

For example, in the sentence "I love pizza", the bigram "love pizza" carries more meaning than the individual words "love" and "pizza". By creating bigrams, we can help the topic model identify that "love pizza" is a distinct concept that should be considered as a single topic.

Creating bigrams in topic models can help to improve the accuracy and relevance of the topics identified by the model.

```python
import numpy as np 

from gsdmm import MovieGroupProcess #GSDMM's other name is "Movie Group Process" (reference to the article first published on GSDMM)

from gensim.models.coherencemodel import CoherenceModel #for coherence score
from gensim import corpora, models
from gensim.utils import simple_preprocess
import gensim, spacy
from gensim.models.ldamulticore import LdaMulticore #LDA 
from gensim.models import Phrases #Bigrams
from gensim.models.phrases import Phraser
```

```python
df['preprocessed'] = df.preprocessed.values.tolist()

def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence)))
```

```python
# create N-grams
def make_n_grams(texts):
    bigram = gensim.models.Phrases(texts, min_count=5, threshold=100)  # higher threshold fewer phrases.
    bigram_mod = gensim.models.phrases.Phraser(bigram)
    trigram = gensim.models.Phrases(bigram[texts], threshold=100)
    trigram_mod = gensim.models.phrases.Phraser(trigram)
    bigrams_text = [bigram_mod[doc] for doc in texts]
    trigrams_text =  [trigram_mod[bigram_mod[doc]] for doc in bigrams_text]
    return trigrams_text
    # setting a lower threshold will result in more phrases being detected, including less frequent ones
    # between 100 and 1000 is often used.
```

## Hyper-parameters of the GSDMM model:


**K = 15** It is the MAXIMUM number of clusters. We set this value after several experiments in which we started from 30 clusters. As we increased the number of clusters, the empty clusters started to appear. This was why we settled with 15 clusters. One might try different K values. 

**Alpha α** — In regards to the MGP, alpha controls a factor that decides how easily a table gets removed when it’s empty (low alpha = fewer tables). As a result, the number of non-empty clusters will get larger as alpha increases and there will be more clusters with only one or few documents. By default, a value of α=0.1 works fine on most datasets.


**Beta β** — In regards to the MGP, beta controls how a table is chosen for a student based on similarity to the table rather than the popularity of the table (low beta = more similar clusters). As a result, the number of non-empty clusters will get smaller as beta increases since the model will give more emphasis on selecting popular clusters rather than similar clusters. By default, a value of β=0.1 works fine on most data sets.


**N_iters** — In regards to the MGP, this represents the number of iterations or number of times that each student is reassigned to a new table by the professor. According to the original research paper, GSDMM converges fairly quickly and stability is generally reached within 10 though we will use n_iters=30as our default.
[Reference](https://stackoverflow.com/questions/62108771/a-practical-example-of-gsdmm-in-python)

**In summary:**

**Low alpha result in many clusters with single words, while high alphas result in less clusters with more words.** 

**High beta (eta) result in popular clusters while low beta result in similar clusters (which are not strong populated). What parameteres you need is based on the dataset. The number of clusters can mostly controlled by beta, but alpha has also (as described) an influence.**

```python
mgp = MovieGroupProcess(K=15, alpha=0.1, beta=0.1, n_iters=30) 

vocab = set(x for review in tokens_reviews for x in review)
n_terms = len(vocab)
model = mgp.fit(tokens_reviews, n_terms)
```

```python
'''iterates over each document in data_text, assigning the preprocessed text 
to the "Text" column of the DataFrame and the corresponding 
lemmatized text to the "Lemma-text" column.'''

def create_topics_dataframe(data_text=df.preprocessed_joined,  mgp=mgp, threshold=0.3, topic_dict=topic_dict, lemma_text=tokens_reviews):
    result = pd.DataFrame(columns=['Topic', 'Lemma-text']) #if the probability of the highest-scoring topic for a document is above the threshold, the document is assigned to that topic.
    
    for i, text in enumerate(data_text):
        result.at[i, 'Text'] = text
        result.at[i, 'Lemma-text'] = lemma_text[i]
        prob = mgp.choose_best_label(tokens_reviews[i])
        if prob[1] >= threshold:
            result.at[i, 'Topic'] = topic_dict[prob[0]]
        else:
            result.at[i, 'Topic'] = 'Other'
            
            
    return result
```

**TF-IDF (Term Frequency-Inverse Document Frequency)** is a statistical measure used to evaluate the importance of a word in a document or corpus. It is based on the idea that words that are common in a document but rare in the overall corpus are more informative and valuable in describing the content of that document. TF-IDF does not remove all frequent and least frequent words from a document or corpus. Instead, it calculates a score for each word based on its frequency in the document and its frequency in the corpus. Words that are common in the document but rare in the corpus will have a high TF-IDF score, while words that are common in both the document and the corpus will have a low TF-IDF score.
Therefore, the words with the highest TF-IDF scores are the ones that are most important in describing the content of a document, while the words with the lowest TF-IDF scores are the ones that are least important. However, TF-IDF does not remove any words from the document or corpus itself. It simply ranks the words based on their importance.

TF-IDF Intuition:

Imagine you have a bunch of stories that you want to analyze to see which words are most important in each story.

TF-IDF looks at how often a word appears in each story (we call that TF or term frequency) and also how important that word is overall (we call that IDF or inverse document frequency).

Then, it combines those two pieces of information to give each word a score that tells us how important it is in that particular story.

When we use TF-IDF on a bunch of stories, we can find out which words are really important in each one, which can help us understand what the story is about and what's most interesting or significant in it.

**Do we need to remove most and least frequent words even before applying TF-IDF?**
In general, it is a good idea to remove the most and least frequent words in a corpus of tweets before applying topic modeling techniques like Latent Dirichlet Allocation (LDA). This is because LDA works by identifying patterns of co-occurrence among words in the corpus, and words that are very common or very rare can introduce noise into the model.

However, the need to remove most and least frequent words depends on the specific characteristics of the corpus and the goals of the analysis. In some cases, using TF-IDF may be sufficient to downweight the importance of very common or very rare words, and removing them may not be necessary.

Since we don't use TF-IDF in GSDMM, we removed the least and most frequent words. 

## Parameters and Topic Numbers 

```python
def calc_coherence_values(dictionary, corpus, texts, limit=30, start=10, step=5):
    coherence_values = [] #empty list for coherence values 
    model_list = [] #empty list to store models 
    hyperparam_list = [] #empty list to store related hyperparameters 
    no_of_topics = list(range(start, limit, step)) #topic numbers between 10-30
    alpha_list = [0.1, 0.3, 'symmetric'] #trying multiple hyperparameters 
    beta_list = [0.01, 'auto']
    
    for num_topics in no_of_topics:
        for alpha in alpha_list:
            for beta in beta_list:
                model = LdaMulticore(corpus=corpus, id2word=dictionary, num_topics=num_topics, alpha=alpha, eta=beta, random_state=42)
                model_list.append(model) #for each topic number, for each hyperparameter, it trains a model and appends the model in the model_list
                hyperparam_list.append(f"num_topics:{num_topics}, alpha:{alpha}, eta:{beta}") #apppend 
                coherence_model = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v') #this particular metric is default and heavily used
                coherence_score = coherence_model.get_coherence()
                coherence_values.append(coherence_score)
                print(f"Num topics: {num_topics}, Alpha: {alpha}, Beta: {beta}, Coherence Score: {coherence_score}")
                
    return model_list, coherence_values, hyperparam_list

model_list, coherence_values, hyperparam_list = calc_coherence_values(dictionary=id2word, corpus=corpus_tfidf, texts=tokens_reviews, limit=30, start=10, step=5)
```

```python
from sklearn.decomposition import NMF

# dictionary and corpus 
id2word = corpora.Dictionary(tokens_reviews) 
texts = tokens_reviews
corpus = [id2word.doc2bow(text) for text in texts] 

doc_term_matrix = gensim.matutils.corpus2csc(corpus).transpose() # convert the corpus into a document-term matrix for NFM
#NFM
num_topics = 10 
nmf_model = NMF(n_components=num_topics, init='random', random_state=0, max_iter=1000) #default settings
nmf_matrix = nmf_model.fit_transform(doc_term_matrix)
```

## NMF Topics 

```python
# top words in each topic
topic_words = nmf_model.components_
for i, topic in enumerate(topic_words):
    word_indices = topic.argsort()[-20:]
    top_words = []
    for index in word_indices:
        word = id2word[index]
        top_words.append(word)
    print(f"Topic {i}: {', '.join(top_words)}")
    print("="*200)

#looks better than LDA or gsdmm for short texts.  
```

**Read some of the documents that are classified under the topic.** This can help to provide context for the words that are associated with the topic and give a better understanding of what the topic might be about.

**Use external knowledge or domain expertise to help interpret the topics.** If you have knowledge about the subject matter being studied, you may be able to infer the topic from the words that are associated with it.

**Consider using visualization tools to help interpret the topics.** Topic modeling software often provides visualization tools that can help to identify patterns in the data and make it easier to interpret the topics.

## 2.5 t-SNE <a class="anchor" id="two-five"></a>


**Documentation**
Code from: [Topic Modelling basics: LDA, LSA and t-SNE](https://www.kaggle.com/code/yash161101/topic-modelling-basics-lda-lsa-and-t-sne)


t-SNE stands for t-Distributed Stochastic Neighbor Embedding. It's a tool in ML to help us visualize high-dimensional data in a lower-dimensional space. Basically, it takes a bunch of data that has many different features, and tries to group similar data points together in a way that makes it easier for us to understand.

**Topic Model Visualiastion:** For example, if we have a topic model that's found five different topics in a collection of news articles, we can use t-SNE to plot those five topics in a way that shows which ones are more closely related to each other.

```python
#!pip install bokeh
from sklearn.manifold import TSNE
from bokeh.plotting import figure, output_file, show
from bokeh.models import Label
from bokeh.io import output_notebook
from matplotlib import pyplot as plt
from matplotlib import colors as mcolors

# get topic weights
topic_weights = [] #stores the topic weights for each document
for row in nmf_model.transform(doc_term_matrix): #transform gets topic weights 
    topic_weights.append([w for w in row]) #iterates over each row 
arr = pd.DataFrame(topic_weights).fillna(0).values #converts the topic_weights into array, as you can see below, sometimes topic weight can be zero 

# Dominant topic number in each document
topic_num = np.argmax(arr, axis=1) #finds the maximum weight value and assigns the topic 

# tSNE 
tsne_model = TSNE(n_components=2, verbose=1, random_state=0, angle=.99, init='pca')
tsne_nmf = tsne_model.fit_transform(arr)

```

```python
print(topic_weights[1])
print("="*200)
print(topic_weights[2])
```

```python
print(topic_num[1])
print("="*200)
print(topic_num[2])
```

```python
# plotting the clusters 
output_notebook()
n_topics = 10
mycolors = np.array([color for name, color in mcolors.TABLEAU_COLORS.items()])
plot = figure(title="{} NMF Topics".format(n_topics), width=1200, height=1000)
for i in range(n_topics):
    plot.scatter(x=tsne_nmf[topic_num==i, 0], y=tsne_nmf[topic_num==i, 1], color=mycolors[i], legend_label='Topic {}'.format(i))

# add labels for each topic
for i in range(n_topics):
    top_words = [id2word[index] for index in nmf_model.components_[i].argsort()[-3:]] #gets top 3 words of each topic
    label = Label(x=tsne_nmf[topic_num==i, 0].mean(),y=tsne_nmf[topic_num==i, 1].mean(),text='Topic {}: {}'.format(i, ', '.join(top_words)),text_font_size='16pt')
    plot.add_layout(label)

plot.legend.click_policy="hide"
show(plot)
```

# Topic Models Pitfalls  <a class="anchor" id="three"></a>


Topic models can sometimes produce **ambiguous or overlapping topics**, where the same word or phrase may appear in multiple topics which can making it difficult to interpret the underlying themes.

Topic models are based on word co-occurrence patterns and may not capture the full context or nuances of language, leading to **inaccurate or incomplete representations of the underlying topics**.

The **quality and representativeness** of the input data can have a significant impact on the quality of the resulting topics, and preprocessing steps such as stemming or stop-word removal can introduce biases into the model.

The resulting topics can be **difficult to interpret or explain**, especially if they are abstract. This can make it challenging to apply the results in practical applications.

You can always use annotated data with topic classification with machine learning or deep learning.
